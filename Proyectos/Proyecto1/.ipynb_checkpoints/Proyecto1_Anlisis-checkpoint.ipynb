# Análisis del Dataset de los Videos de Youtube

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Cargar el archivo

```python
path_youtube = "C:\\Users\\ro199\\Documents\\GitHub\\py-alvarado-roman-ronald-francisco\\Proyectos\\Proyecto1\\CAvideos.csv"
```

## Lectura del archivo que contine 40881 filas y 16 columnas

```python
youtube_dataset = pd.read_csv(path_youtube)
print(youtube_dataset.shape)
```

## Obtener los titulos de las columnas del archivo

```python
Todas_columnas = youtube_dataset.columns.values
print(Todas_columnas)
```

## Mostrar las 5 primeras filas del archivo

```python
display(youtube_dataset.head(5))
```

# Primer Ejemplo
## Top de las 100 videos mas vistos de acuerdo con el numero de views


```python
top100_videos_mas_vistos = youtube_dataset['views'].sort_values(ascending=False).head(100)
videos_mas_vistos = youtube_dataset.loc[top100_videos_mas_vistos.index,['video_id','trending_date','title','channel_title','category_id','tags','views','likes','dislikes','comment_count']]
print('TOP 100:')
display(videos_mas_vistos)
```

# Segundo Ejemplo
## Top de las 100 videos menos vistos de acuerdo con el numero de views


```python
top100_videos_menos_vistos = youtube_dataset['views'].sort_values(ascending=True).head(100)
videos_menos_vistos = youtube_dataset.loc[top100_videos_menos_vistos.index,['video_id','trending_date','title','channel_title','category_id','publish_time','tags','views']]
print('TOP 100:')
display(videos_menos_vistos)
```

```python
videos_menos_vistos.groupby(['category_id'])['views'].min().plot(kind='line', title="Videos menos vistos")
```

# Tercer Ejemplo
## Top de las 20 videos que no fueron muy aceptados por los usarios de youtube de acuerdo con el numero de dislikes


```python
top20_videos_dislike = youtube_dataset['dislikes'].sort_values(ascending=False).head(20)
videos_mas_dislikes = youtube_dataset.loc[top20_videos_dislike.index,['trending_date','title','channel_title','publish_time','tags','dislikes', 'category_id']]
print('TOP 20:')
display(videos_mas_dislikes)
```

```python

my_plot = youtube_dataset.plot("category_id", "dislikes", kind="scatter")
plt.show()
```

# Cuarto Ejemplo
## Top de las 20 videos que fueron muy aceptados por los usarios de youtube de acuerdo con el numero de likes

```python
top20_videos_like = youtube_dataset['likes'].sort_values(ascending=False).head(20)
videos_mas_likes = youtube_dataset.loc[top20_videos_like.index,['title','category_id','likes','dislikes']]
print('TOP 20:')
display(videos_mas_likes)
```

```python
videos_mas_likes.groupby('likes', 'dislikes').count()["category_id"].plot(kind='bar',stacked=True,title="Total de videos con likes vs likes")
```

# Quinto Ejemplo
## Porcentajes de los videos mas vistos con sus repectivos titulos por los usuarios de youtube 

```python
porcentaje_vistos_titulo = videos_mas_vistos['title']
print('Procentaje de videos mas vistos')
display(porcentaje_vistos_titulo.value_counts(normalize=True))
```

# Sexto Ejemplo
## Porcentajes de los videos con mas likes por los usuarios de youtube 

```python
porcentaje_likes_titulo = videos_mas_likes['title']
print('Procentaje de videos con mas likes')
display(porcentaje_likes_titulo.value_counts(normalize=True))
name = []
for i in porcentaje_likes_titulo.value_counts():
    name.append(i)
values = [0.25,0.25,0.20,0.15,0.10,0.05]
plt.pie(values, labels=name)
plt.axis("equal")
plt.show()
```

# Septimo Ejemplo
## Top de las 20 videos que fueron muy comentados por los usarios de youtube de acuerdo con el numero de likes

```python
top20_videos_comment = youtube_dataset['comment_count'].sort_values(ascending=False).head(20)
videos_mas_comentados = youtube_dataset.loc[top20_videos_comment.index,['trending_date','title','channel_title','comment_count']]
print('TOP 20:')
display(videos_mas_comentados)
```

# Octavo Ejemplo
## Porcentajes de los videos mas comentados por los usuarios de youtube 

```python
porcentaje_comment_title = videos_mas_comentados['title']
print('Procentaje de videos mas comentados')
display(porcentaje_comment_title.value_counts(normalize=True))
```
