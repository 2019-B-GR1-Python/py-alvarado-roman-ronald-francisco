## 1. Introduccion
### La plataforma de youtube es un repositorio de videos streaming y de entretenimiento hacia sus usuarios.
### Su objetivo rpincipal es dar voz a todas las personas y poner el mundo a su alcance.
## 2. Descripcion
### YouTube es un portal del Internet que permite a sus usuarios subir y visualizar videos. Fue creado en febrero de 2005 por Chad Hurley, Steve Chen y Jawed Karim, quienes se conocieron trabajando en PayPal. Un a√±o m√°s tarde, YouTube fue adquirido por Google en 1.650 millones de d√≥lares.

### Esta plataforma cuenta con un reproductor online basado en Flash, el formato desarrollado por Adobe Systems. Una de sus principales innovaciones fue la facilidad para visualizar videos en streaming, es decir, sin necesidad de descargar el archivo a la computadora. Los usuarios, por lo tanto, pueden seleccionar qu√© video quieren ver y reproducirlo al instante.
## 3. Desarrollo
### Para realizar el siguiente an√°lisis sobre la plataforma de YouTube  se obtuvo el dataset de la pagina Kaggle, la cual est√° conformada por 16 columnas, las cuales ser√°n descritas a continuaci√≥n:
#### 'video_id' = Numero unico que identifica el video
#### 'trending_date' = Fecha en el que el video se hizo popular
####  'title' = Nombre del video
#### 'channel_title' =  Canal que publico el video
####  'category_id' = Numero unico que identifica la categoria
#### 'publish_time' = Fecha y hora de publicacion 
####  'tags' = Etiquetas que identifican al video
####  'views' = Numero de vistas que presenta el video
#### 'likes' = Numero de "Me gustas" que presenta el video
####  'dislikes' = Numero de "No me gustas" que presenta el video
####  'comment_count' = Numero de "Comentarios" que presenta el video
#### 'thumbnail_link' = Direccion url del video
#### 'comments_disabled' = Si es false muetra que se bloqueo el comentario
#### 'ratings_disabled' =  = Si es false muetra que no hubo error
#### 'video_error_or_removed' = Si es false muetra que no hubo error
#### 'description' = Pequenia descripcion del que publica el video

# Primer Ejemplo
## Top de las 100 videos mas vistos de acuerdo con el numero de views


```python
top100_videos_mas_vistos = youtube_dataset['views'].sort_values(ascending=False).head(100)
videos_mas_vistos = youtube_dataset.loc[top100_videos_mas_vistos.index,['video_id','trending_date','title','channel_title','category_id','tags','views','likes','dislikes','comment_count']]
print('TOP 100:')
display(videos_mas_vistos)
```

```python
youtube_dataset.groupby(['category_id'])['views'].describe().plot(kind='line', title = "Grafica descriptiva por su categoria")
```

# Segundo Ejemplo
## Top de las 100 videos menos vistos de acuerdo con el numero de views


```python
top100_videos_menos_vistos = youtube_dataset['views'].sort_values(ascending=True).head(10)
videos_menos_vistos = youtube_dataset.loc[top100_videos_menos_vistos.index,['video_id','trending_date','title','channel_title','category_id','publish_time','tags','views']]
print('TOP 100:')
display(videos_menos_vistos)
```

```python
videos_menos_vistos.groupby(['category_id'])['views'].min().plot(kind='line', title="Videos menos vistos")
```

# Cuarto Ejemplo
## Top de las 20 videos que fueron muy aceptados por los usarios de youtube de acuerdo con el numero de likes

```python
top20_videos_like = youtube_dataset['likes'].sort_values(ascending=False).head(20)
videos_mas_likes = youtube_dataset.loc[top20_videos_like.index,['title','category_id','likes','dislikes']]
print('TOP 20:')
display(videos_mas_likes)
```

# Quinto Ejemplo
## Porcentajes de los videos mas vistos con sus repectivos titulos por los usuarios de youtube 

```python
porcentaje_vistos_titulo = videos_mas_vistos['title']
print('Procentaje de videos mas vistos')
display(porcentaje_vistos_titulo.value_counts(normalize=True))
```

# Sexto Ejemplo
## Porcentajes de los videos con mas likes por los usuarios de youtube 

```python
porcentaje_likes_titulo = videos_mas_likes['title']
print('Procentaje de videos con mas likes')
display(porcentaje_likes_titulo.value_counts(normalize=True))
name = []
for i in porcentaje_likes_titulo.value_counts():
    name.append(i)
values = [0.25,0.25,0.20,0.15,0.10,0.05]
plt.pie(values, labels=name)
plt.axis("equal")
plt.show()
```

```python
import pylab as pl
ax = pl.axes([0.025, 0.025, 0.95, 0.95], polar=True)

N = 20
theta = np.arange(0.0, 2 * np.pi, 2 * np.pi / N)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * 10
bars = pl.bar(theta, radii, width=width, bottom=0.0)

for r,bar in zip(radii, bars):
    bar.set_facecolor(pl.cm.jet(r/10.))
    bar.set_alpha(0.5)

ax.set_yticklabels([0.05,0.15,0.20,0.25,0.35])
pl.show()
```

# Doceavo Ejemplo
## Grafica comentarios desactivados del acuerdo al numero de visitas

```python
plt.figure(figsize = (20,7))
sns.boxplot(  y = youtube_dataset['views'] ,x = youtube_dataset['comments_disabled'])
plt.grid(True)
plt.title('Comentarios desactivados por el numero vistas que posee el video del Canal')
plt.show()
```

```python
name1 = []
for i in porcentaje_comment_title.value_counts():
    name1.append(i)
plt.figure(figsize = (18,7))
plt.plot(porcentaje_comment_title.value_counts() , name1, linestyle = '-', marker = 'o', markerfacecolor = 'red')
plt.grid(True)
plt.xlabel('Numero de Vistas')
plt.ylabel('Videos con errores o removidos')
plt.title('Comentarios desactivados por el numero vistas que posee el video del Canal')
plt.show()
```
